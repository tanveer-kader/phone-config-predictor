# Phone Price Segmentation

### Import all Dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

### Data Loading and Preprocessing

```python
# Load dataset
mobile_dataset = pd.read_csv("train.csv")
```

```python
# show dataset
mobile_dataset.head()
```

```python
# dataset size
mobile_dataset.shape
```

```python
# dataset information
mobile_dataset.info()
```

```python
# checking missing values
mobile_dataset.isnull().sum()
```

```python
# Price segments and data distribution
mobile_dataset["price_range"].value_counts()
```

```python
# statistical measures of dataset
mobile_dataset.describe()
```

```python
# correlation matrix
correlation = mobile_dataset.corr()
correlation
```

```python
# correlation in heatmap
plt.figure(figsize = (15,15))
sns.heatmap(data = correlation, fmt = "0.2f", cbar = True, square = True, annot = True, annot_kws = {"size": 10})
```

```python
# separating feature and target
X_train = mobile_dataset.drop(["price_range"], axis = 1)
Y_train = mobile_dataset["price_range"]
```

### Data standardisation

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
standardised_data = scaler.transform(X_train)
```

```python
standardised_data
```

```python
# updating X with standardised data
X_train = standardised_data
```

```python
# load test data
test_data = pd.read_csv("test.csv")
test_data.head()
```

```python
test_data = test_data.drop(["id"], axis = 1)
```

```python
# standardise test data
scaler.fit(test_data)
X_test = scaler.transform(test_data)
```

```python
print("X Train data: ", X_train.shape)
print("X Test data: ", X_test.shape)
print("Y Train data: ", Y_train.shape) 
```

### Model Training with Random Forest Classifier

```python
model_rf = RandomForestClassifier()
```

```python
# train model
model_rf.fit(X_train, Y_train)
```

### Model Evaluation

```python
# score on train data
X_train_prediction = model_rf.predict(X_train)
X_train_accuracy = accuracy_score(X_train_prediction, Y_train)
print("Accuracy on training data: ", X_train_accuracy)
```

```python
# plot prediction
plt.scatter(Y_train, X_train_prediction)
plt.title("Actual vs Predicted Price Segment")
plt.xlabel("Actual Segment")
plt.ylabel("Predicted Segment")
```

### Model Training with Support Vector Classifier

```python
model_svm = SVC()
```

```python
# train model
model_svm.fit(X_train, Y_train)
```

### Model Evaluation

```python
# score on train data
X_train_prediction = model_svm.predict(X_train)
X_train_accuracy = accuracy_score(X_train_prediction, Y_train)
print("Accuracy on training data: ", X_train_accuracy)
```

```python
# plot prediction
plt.scatter(Y_train, X_train_prediction)
plt.title("Actual vs Predicted Price Segment")
plt.xlabel("Actual Segment")
plt.ylabel("Predicted Segment")
```

### Prediction on test data

```python
# Predict on test data
rf_prediction = model_rf.predict(X_test)
svm_prediciton = model_svm.predict(X_test)
```

```python
# prediction comparison of models
plt.scatter(rf_prediction, svm_prediciton)
plt.title("RF vs SVM Price Segment")
plt.xlabel("RF Prediction")
plt.ylabel("SVM Prediction")
```
